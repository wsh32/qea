```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x, y = np.meshgrid(np.linspace(-15, 15, 100), np.linspace(-15, 15, 100))
```

```python
f = (x-1)**2 + (y-2)**2
plt.pcolormesh(x, y, f)
plt.axis('equal')
```

```python
x, y = np.meshgrid(np.linspace(-5, 5, 10), np.linspace(-5, 5, 10))
u = x**2 - y
v = x + y**2
plt.quiver(x, y, u , v)
```

```python
x, y, z = np.meshgrid(np.linspace(-5, 5, 5), np.linspace(-5, 5, 5), np.linspace(-5, 5, 5))
u = 2*y - z
v = x + y**2 - z
w = 4*y - 3*x

fig = plt.figure()
ax = fig.gca(projection='3d')
ax.quiver(x, y, z, u, v, w, normalize=True)
```

```python
x, y = np.meshgrid(np.linspace(-30, 30, 100), np.linspace(-25, 25, 100))
V = np.log(x**2 + y**2)
plt.contour(x, y, V)
plt.axis('equal')
```

```python
xg, yg = np.meshgrid(np.linspace(-25, 25, 6), np.linspace(-25, 25, 6))
px = xg / (xg**2 + yg**2)
py = yg / (xg**2 + yg**2)
plt.quiver(xg, yg, px, py)
plt.contour(x, y, V)
plt.axis('equal')
```

```python
x, y = np.meshgrid(np.linspace(-5, 5, 100), np.linspace(-5, 5, 100))
V = np.real(np.log(x**2 + y**2) - np.log((x-1)**2 + y**2))
plt.contourf(x, y, V, np.linspace(-1, 1, 50), cmap='RdBu')
plt.xlim([-5, 5])
plt.ylim([-5, 5])
```

```python
xg, yg = np.meshgrid(np.linspace(-5, 5, 8), np.linspace(-5, 5, 8))
px = 2*xg / (xg**2 + yg**2) - (2*xg - 2) / ((xg-1)**2 + yg**2)
py = 2*yg / (xg**2 + yg**2) - 2*yg / ((xg-1)**2 +yg**2)
plt.contourf(x, y, V, np.linspace(-1, 1, 50), cmap='RdBu')
plt.quiver(xg, yg, px, py)
plt.xlim([-5, 5])
plt.ylim([-5, 5])
plt.axis('equal')
```

```python
V = np.real(np.log(x**2 + y**2) - np.log((x-1)**2 + y**2) - np.log((x-2)**2 + (y-3)**2))
xg, yg = np.meshgrid(np.linspace(-5, 5, 10), np.linspace(-5, 5, 10))
px = 2*xg / (xg**2 + yg**2) - (2*xg - 2) / ((xg-1)**2 + yg**2)
py = 2*yg / (xg**2 + yg**2) - 2*yg / ((xg-1)**2 +yg**2)
plt.contourf(x, y, V, np.linspace(-10, 8, 50), cmap='RdBu')
plt.quiver(xg, yg, px, py)
plt.xlim([-5, 5])
plt.ylim([-5, 5])
plt.axis('equal')
```
